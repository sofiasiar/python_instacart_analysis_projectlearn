```python
#Create function price_label
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Define function price_label 
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Define the conditions for the busiest periods of the day
def categorize_hour(hour):
    if hour in range(10, 18):  # Midday peak hours
        return "Most orders"
    elif hour in range(6, 10) or hour in range(18, 22):  # Morning & evening moderate orders
        return "Average orders"
    else:  # Early morning and late night
        return "Fewest orders"
```

```python
# Apply function to create the new column
df["busiest_period_of_day"] = df["order_hour_of_day"].apply(categorize_hour)
```
